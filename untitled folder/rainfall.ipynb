# Methodology

- Discard rainfall events that happens during November - May
- Construct a data base of the rainfall features (duration, precipitation, intensity)

## Assumptions
- A new event starts if there is at least one dry hour (0 mm rainfall) between two rainfall periods.


```python
import numpy as np
import pandas as pd
from pathlib import Path
from fitter import Fitter
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ExpSineSquared, WhiteKernel, ConstantKernel
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

## Parameters

```python
data_folder = "./data"

# winter months
winter_months = [11, 12, 1, 2, 3, 4]

IETD_threshold = 6
```

## Methods

### Data Preprocessing
- Concatenate months
- Rainfall Events Discretization

```python
def concat_data(data_folder, year):
    """
    Concatenate CSV files from a specified folder and year into a single DataFrame.

    Args:
        data_folder (str): Path to the folder containing the CSV files.
        year (int): Year to filter the CSV files by.

    Returns:
        pd.DataFrame: DataFrame containing concatenated data from the specified year.
    """
    
    data = pd.DataFrame()
    for file in Path(data_folder).rglob('*.csv'):
        # Check if the year number is found in the file name
        if any(str(year) in file.name):  # Adjust the range as needed
            print(file)
            df = pd.read_csv(file)
            data = pd.concat([data, df])
    return data
```

```python
def extract_rainfall_events(df, time_col, rain_col, winter_months=winter_months):
    """
    Extracts rainfall events from hourly precipitation data while discarding events from November to April.

    Parameters:
        df (pd.DataFrame): DataFrame containing precipitation data.
        time_col (str): Name of the column containing timestamps.
        rain_col (str): Name of the column containing hourly precipitation values.

    Returns:
        pd.DataFrame: DataFrame containing rainfall events with features (duration, precipitation, intensity).
    """
    
    # Convert time column to datetime if it's not already
    df[time_col] = pd.to_datetime(df[time_col])

    # Filter data to only include months May - October
    df = df[~df[time_col].dt.month.isin(winter_months)]

    # Sort data by time in case it's not ordered
    df = df.sort_values(by=time_col).reset_index(drop=True)

    # Identify rainfall events (separated by at least one dry hour)
    events = []
    event_start = None
    total_precip = 0
    peak_precip = 0
    duration = 0
    IETD = 0
    
    for i in range(len(df)):
        rain = df[rain_col].iloc[i]
        timestamp = df[time_col].iloc[i]

        if rain > 0:  # Rainfall detected
            IETD = 0
            if event_start is None:  # Start of a new event
                event_start = timestamp
                duration = 1
                total_precip = rain
            else:  # Ongoing event
                duration += 1
                total_precip += rain
                peak_precip = max(peak_precip, rain)
        else:  # Dry hour (potential event break)
            IETD += 1
            if event_start is not None and IETD >= IETD_threshold:  # If we were tracking an event
                events.append({
                    "Start Time": event_start,
                    "End Time": df[time_col].iloc[i - 1],
                    "Duration (hrs)": duration,
                    "Total Precipitation (mm)": total_precip,
                    "Intensity (mm/hr)": total_precip / duration,
                    "Peak Precipitation (mm)": peak_precip
                })
                event_start = None  # Reset event tracking
                duration = 0
                total_precip = 0
                

    # Convert events into a DataFrame
    event_df = pd.DataFrame(events)

    return event_df
```

```python
def winter_month(df, time_col, winter_months=winter_months):
    """
    Check if winter month from the DataFrame.

    Parameters:
        df (pd.DataFrame): DataFrame containing precipitation data.
        time_col (str): Name of the column containing timestamps.

    Returns:
        bool: True if the month is between November - April, False otherwise.
    """
    # Convert time column to datetime if it's not already
    df[time_col] = pd.to_datetime(df[time_col])

    # check if month is between November - April
    it_is = np.all(df[time_col].dt.month.isin(winter_months))

    return it_is
```

```python
def extract_inter_event_times(event_df):
    """
    Extracts inter-event times from event data.

    Parameters:
        event_df (pd.DataFrame): DataFrame containing event data with start and end times.

    Returns:
        pd.Series: Series containing inter-event times in hours.
    """
    
    # Calculate inter-event times
    inter_event_times = (event_df["Start Time"] - event_df["End Time"].shift(1)).dt.total_seconds() / 3600
    # inter_event_times = inter_event_times[:-1]  # Remove last NaN value (due to shift)

    # Create the dataframe
    inter_event_times_df = pd.DataFrame({
        "1st Event - End Time": event_df["End Time"].shift(1),
        "2nd Event - Start Time": event_df["Start Time"],
        "Inter-Event Time (hrs)": inter_event_times
        })
    
    # Skip inter-event times between months in different years
    inter_event_times_df = inter_event_times_df[
        (inter_event_times_df["1st Event - End Time"].dt.year == 
         inter_event_times_df["2nd Event - Start Time"].dt.year)
    ]

    # # Edit events that work inter years so the event in next year is considered from starting from its month
    # inter_event_times_df.loc[
    #     (inter_event_times_df["1st Event - End Time"].dt.year != 
    #         inter_event_times_df["2nd Event - Start Time"].dt.year), 
    #     "Inter-Event Time (hrs)"
    # ] = inter_event_times_df["2nd Event - Start Time"].apply(
    #     lambda x: (x - x.replace(day=1, hour=0, minute=0, second=0, microsecond=0)).total_seconds() / 3600
    #     )


    return inter_event_times_df
```

### Distribution Fitting

```python
from fitter import Fitter
import seaborn as sns
import numpy as np

def fit_best_distribution(data, column_name):
    """
    Fit multiple distributions to the data and find the best fit.

    Parameters:
        data (pd.DataFrame): DataFrame containing the data.
        column_name (str): Name of the column to fit the distributions to.

    Returns:
        dict: Best fitting distribution and its parameters.
    """
    import scipy.stats as st
    import matplotlib.pyplot as plt

    # Fit multiple distributions and find the best fit
    f = Fitter(data[column_name].values, bins=50, distributions=['norm', 'expon', 'gamma', 'lognorm', 'chi2'])
    f.fit()

    # Print the best-fitting distribution and parameters
    best_fit = f.get_best(method="sumsquare_error")
    best_dist_name = list(best_fit.keys())[0]  # Get best distribution name
    best_params = list(best_fit.values())[0]  # Get best distribution parameters
    
    return f, best_fit
```

### Machine Learning

```python
def train_gaussian_process(event_data):
    """
    Train a Gaussian Process Regressor on rainfall event data and evaluate its performance.

    Parameters:
        event_data (pd.DataFrame): DataFrame containing rainfall event data with features (duration, precipitation, intensity).

    Returns:
        GaussianProcessRegressor: Trained Gaussian Process Regressor model.
    """
    # Create a time index (e.g., using 'Start Time')
    event_data["time_index"] = (event_data["Start Time"] - event_data["Start Time"].min()).dt.total_seconds() / 3600  # Convert to hours

    # Select Features (X) and Target (y)
    X = event_data[["time_index"]].values  # Time index as input
    y = event_data["Total Precipitation (mm)"].values  # Predicting total precipitation

    # Normalize target values for better model performance
    scaler = MinMaxScaler()
    y_scaled = scaler.fit_transform(y.reshape(-1, 1)).flatten()

    # Define Gaussian Process kernel (captures trend & periodicity)
    kernel = (
        RBF(length_scale=50.0)  # Captures long-term trends
        + ExpSineSquared(length_scale=1.0, periodicity=24.0)  # Captures daily cycles
        + WhiteKernel(noise_level=0.1)  # Accounts for measurement noise
    )

    # Initialize Gaussian Process Regressor
    gp = GaussianProcessRegressor(kernel=kernel, alpha=0.1, n_restarts_optimizer=10)
    gp.fit(X, y_scaled)

    # Forecast Future Rainfall Events (Next 48 Hours)
    future_hours = 48
    future_time_index = np.arange(event_data["time_index"].max() + 1, event_data["time_index"].max() + future_hours + 1).reshape(-1, 1)

    # Predict future precipitation
    y_pred_scaled, sigma = gp.predict(future_time_index, return_std=True)

    # Convert predictions back to original scale
    y_pred = scaler.inverse_transform(y_pred_scaled.reshape(-1, 1)).flatten()

    # Convert time index to real timestamps
    future_dates = event_data["Start Time"].max() + pd.to_timedelta(np.arange(1, future_hours + 1), unit="h")

    # Plot Observed vs Predicted Rainfall
    plt.figure(figsize=(10, 5))
    plt.plot(event_data["Start Time"], event_data["Total Precipitation (mm)"], label="Observed Rainfall", color="blue")
    plt.plot(future_dates, y_pred, 'r-', label="Forecasted Rainfall")
    plt.fill_between(future_dates, y_pred - 2 * sigma, y_pred + 2 * sigma, color="red", alpha=0.2, label="Uncertainty (±2σ)")
    plt.xlabel("Time")
    plt.ylabel("Total Precipitation (mm)")
    plt.title("Gaussian Process Rainfall Forecasting")
    plt.legend()
    plt.show()

    # Example Prediction for the Next Event
    next_event_time = np.array([[event_data["time_index"].max() + 1]])
    predicted_precipitation, predicted_sigma = gp.predict(next_event_time, return_std=True)
    predicted_precipitation = scaler.inverse_transform(predicted_precipitation.reshape(-1, 1)).flatten()

    print(f"Predicted Rainfall for Next Event: {predicted_precipitation[0]:.2f} mm ± {2*predicted_sigma[0]:.2f} mm")

    return gp
```

```python
# Initialize event_data as an empty DataFrame
event_data = pd.DataFrame()

for file in Path(data_folder).glob("*.csv"):
    
    data = pd.read_csv(file)

    if winter_month(data, "Date/Time (LST)"):
        continue
    else:
        new_data = extract_rainfall_events(data, time_col="Date/Time (LST)", rain_col="Precip. Amount (mm)")
        event_data = pd.concat([event_data, new_data]).drop_duplicates().reset_index(drop=True)

# Sort events by start time
event_data = event_data.sort_values(by="Start Time").reset_index(drop=True)

# Extract inter-event times
inter_event_times = extract_inter_event_times(event_data)
```

```python
# Merge event_data with inter_event_times to add the inter-event time column
event_data = event_data.merge(inter_event_times[['2nd Event - Start Time', 'Inter-Event Time (hrs)']],
                              left_on='Start Time', right_on='2nd Event - Start Time', how='left')

# Drop the redundant '1st Event - End Time' column
event_data = event_data.drop(columns=['2nd Event - Start Time'])

# Display the updated event_data DataFrame
event_data
```
