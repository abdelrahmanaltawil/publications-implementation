# Energy System Results Visualization

Visualizes results from the energy optimization (`src/models/energy.py`, E1–E13)
and energy simulation (`src/helpers/energy_simulation.py`) for a given run.

**Usage:** set `run_dir` to the `data/results/run_YYYYMMDD_HHMMSS/` directory
produced by `src/workflow.py` with `energy: true` in `data/inputs/config.yaml`.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import json

plt.style.use('../data/inputs/plotting.mplstyle')

# ── Point to a specific run ────────────────────────────────────────────────
run_dir = Path('../data/results/run_YYYYMMDD_HHMMSS')   # ← edit this
energy_dir = run_dir / 'energy'

with open(run_dir / 'summary.json') as f:
    summary = json.load(f)
print(summary)
```

```python
# ── Dispatch ────────────────────────────────────────────────────────────────
dispatch = pd.read_csv(energy_dir / 'dispatch.csv')
dispatch.head()
```

```python
# ── State of charge ─────────────────────────────────────────────────────────
soc = pd.read_csv(energy_dir / 'soc.csv')
soc.head()
```

```python
# ── Voltages ────────────────────────────────────────────────────────────────
voltages = pd.read_csv(energy_dir / 'voltages.csv')
voltages.head()
```
